```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/NPC_Behavior_Dataset.csv')
df
```

```python
# Drop the 'Favorite_Phrase' column
df = df.drop('Favorite_Phrase', axis=1)
```

```python
df
```

```python
df.set_index('ID', inplace=True)
```

```python
df
```

```python
x = df.iloc[:,[0,1,2]].to_numpy()
y = df.iloc[:,[3]].to_numpy()
```

```python
from sklearn import preprocessing

std_scaler = preprocessing.StandardScaler()
x_std = std_scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_std,y, test_size=0.2, random_state= 42)
```

```python
from sklearn import linear_model
regressor = linear_model.LinearRegression()
regressor.fit(x_train,y_train)
print("the coefficient is : ", regressor.coef_[0])
print("the intercept is : ", regressor.intercept_)
```

```python
mean  = std_scaler.mean_
ecrtype = np.sqrt(std_scaler.var_)
```

```python
coef_original = regressor.coef_[0] / ecrtype
intercept_original = regressor.intercept_ - np.sum((mean * regressor.coef_[0]) / ecrtype)
```

```python
print("the original coeficient : " , coef_original)
print("the original intercept : ", intercept_original)
```
