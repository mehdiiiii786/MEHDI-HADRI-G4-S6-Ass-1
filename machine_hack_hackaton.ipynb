```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_train_scaled,y_train)
```

```python
from sklearn import linear_model
regressor = linear_model.LinearRegression()
regressor.fit(x_train,y_train)
```

```python
print("the coefficient is : ", regressor.coef_[0])
print("the intercept is : ", regressor.intercept_)
```

```python
import numpy as np
mean  = scaler.mean_
ecrtype = np.sqrt(scaler.var_)
```

```python
coef_original = regressor.coef_[0] / ecrtype
intercept_original = regressor.intercept_ - np.sum((mean * regressor.coef_[0]) / ecrtype)
```

```python
print("the original coeficient : " , coef_original)
print("the original intercept : ", intercept_original)
```

```python
predictions = regressor.predict(x_test)
```

```python
predictions
```

```python
submission = pd.DataFrame({"Predicted electricity cost" : predictions})
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print("R² Score (test set):", r2_score(y_test, predictions))
```

```python
submission.to_csv("submission", index = False, sep = ';')
```
