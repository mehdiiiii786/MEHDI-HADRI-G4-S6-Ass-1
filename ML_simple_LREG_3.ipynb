```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/simplelinearregression.csv')
df
```

```python
plt.scatter(df['Age'], df['Premium'], color = 'blue')
plt.xlabel('Age')
plt.ylabel('prenium')
plt.show()
```

```python
x = df.Age.to_numpy()
y = df.Premium.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train.reshape(-1,1),y_train)
print('the coefficient is : ', regressor.coef_[0])
print('the intercept is : ', regressor.intercept_)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train, regressor.coef_[0]*x_train + regressor.intercept_)
plt.show()
```

```python
y_test_ = regressor.predict(x_test.reshape(-1,1))
y_test_
```

```python
new_df = pd.DataFrame({'Actual_Premium': y_test, 'Predicted_Premium': y_test_})
```

```python
new_df
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
print('r2 is :', r2_score(y_test_,y_test))
```
