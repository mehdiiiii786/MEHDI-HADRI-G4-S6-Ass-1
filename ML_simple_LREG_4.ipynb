```python
first_symbol = df['symbol'].iloc[0]
newdf = df[df['symbol'] == first_symbol]
```

```python
newdf
```

```python
plt.scatter(newdf['open'],newdf['close'])

plt.ylabel('open price')
plt.xlabel('close price')
plt.show()
```

```python
x= newdf.open.to_numpy()
y= newdf.close.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train.reshape(-1,1),y_train)
print('the coefficient is : ',  regressor.coef_[0])
print('the intercept is : ', regressor.intercept_)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train, regressor.coef_[0] * x_train + regressor.intercept_)
```

```python
y_test_ = regressor.predict(x_test.reshape(-1,1))
```
